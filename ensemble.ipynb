```
import statistics

def readFileAsDict(path_file):
    new_dict = {}
    
    with open(path_file) as f:
        for line in f:
            (key, val) = line.split()
            new_dict[key] = int(val)

    return new_dict

def calcMode(mod1, mod2, mod3):
    out_mod = {}

    for key, value in mod1.items():
        out_mod.update({
            key: statistics.mode( [ mod1[key], mod2[key], mod3[key] ] )
        })

    return out_mod


def writeOutFile(path, out):
    file = open(path, 'w')

    for key, value in out.items():
        file.write('{} {}'.format(key, value) + '\n') 

    file.close()


modelo_1 = readFileAsDict('./modelo_1.txt')
modelo_2 = readFileAsDict('./modelo_2.txt')
modelo_3 = readFileAsDict('./modelo_3.txt')

out = calcMode(modelo_1, modelo_2, modelo_3)

writeOutFile('CristianoFilho.txt', out)
```
